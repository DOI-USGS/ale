```python
fileName = '/Users/astamile/testData/ale/ody/R79520005EDR-isis-spiced.cub'

import os
os.environ["ISISDATA"] = "/Users/astamile/isisdata/isis_data"
os.environ["ISISTESTDATA"] = "/Users/astamile/isisdata/isis_testData"
os.environ["ISISROOT"] = "/Users/astamile/ISIS3/build"

import ale 
from ale.drivers.ody_drivers import OdyThemisIrIsisLabelNaifSpiceDriver
from ale.formatters.formatter import to_isd
import spiceypy as spice
from ale.drivers import AleJsonEncoder
```

```python
from ale.util import generate_kernels_from_cube
kernels = generate_kernels_from_cube(fileName, expand=True, format_as='list')
```

```python
with OdyThemisIrIsisLabelNaifSpiceDriver(fileName, props = {"kernels": kernels}) as driver:
    print(driver.ephemeris_start_time)
    isisString = to_isd(driver)
```

```python
import json

isis_dict = isisString

json_file = os.path.splitext(fileName)[0] + '.json'

with open(json_file, 'w') as fp:
    json.dump(isis_dict, fp, cls = AleJsonEncoder)
    
with open(json_file, 'r') as fp:
    isis_dict = json.load(fp)
    
isis_dict.keys()
```
