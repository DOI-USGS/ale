```python
import ale
import os
from ale.formatters.formatter import to_isd

os.environ['ISISDATA'] = '/Volumes/usgs-data/isis-data/'

# Set to the directory with your cubes
working_dir = '/Users/username/data/'

# These 2 MRO cubes were found on MROX_0003 in the PDS Data Volumes hosted by USGS.
# https://pdsimage2.wr.usgs.gov/Mars_Reconnaissance_Orbiter/CTX/mrox_0003/data/
cube1 = 'T01_000801_1353_XN_44S195W.cub'
cube2 = 'T01_000801_1867_XN_06N202W.cub'
```

```python
loc1 = working_dir + cube1
loc2 = working_dir + cube2

# Using Cube 1, find the driver
my_driver = ale.drivers.load_driver(loc1)

# Use the driver on Cube 2
with my_driver(loc2) as driver:
    my_isd = to_isd(driver)

# Print ISD of Cube 2 for inspection.
my_isd
```
